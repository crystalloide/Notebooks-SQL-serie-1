La clause WHERE est une clause facultative de l'instruction SELECT. 

Elle apparaît après la clause FROM comme dans l'instruction suivante :

***SELECT column_list FROM table_name WHERE search_condition;***

### 2. Récupération des données avec WHERE

Prenons ici en exemple la table ***rch***

#### 2.1 Et allons vérifier tout d'abord les colonnes qui composent cette table.

```python
%sql SELECT * From rch LIMIT 5
```

#### 2.2 Vérifions le nombre de lignes de la table

Il devrait y avoir 8280 lignes. 

Cela peut être vérifié la fonction dénommée ***COUNT***. 

Nous aborderons d'autres fonctions de SQLite dans les prochains notebooks.

```python
%sql SELECT COUNT(*) as nrow From rch
```

#### 2.3 Utilisons la clause ***WHERE*** pour filter les données à récupérer

Disons que nous ne nous intéressons qu'aux enregistrements de l'année 1981, via le champ ***YR***. 

L'utilisation d'un WHERE est assez simple dans un cas comme celui-ci.

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR=1981
```

#### 2.4 Utilisons maintenant la clause ***AND*** pour filtrer un peu plus les données :


Il y a 23 lignes retournées dans notre requête précédente sur la table RCH. 

Mais disons que nous ne sommes finalement intéressés que les lignes pour lesquelles le champ (colonne) RCH est égal à 10. 

Nous pouvons ajouter une autre condition de filtre avec une instruction ***AND***.

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR=1981 AND RCH=10
```

#### 2.5 Essayons encore plus de combinaison de filtres :

Nous pouvons également filtrer davantage les données avec les opérateurs *!=* ou **<> **

pour obtenir les données avec RCH = 10 et MO=6 mais pas celles concernant "1981".


```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR<>1981 and RCH=10 and MO=6
```

Nous pouvons filtrer davantage les données sur des mois spécifiques à l'aide de l'instruction ***OR***. 

Par exemple, nous aimerions vérifier les données pour les mois de 3, 6 et 9. 

Et nous pourrions utiliser ***()*** pour en faire une condition. 

C'est une astuce à essayer !

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR>2009 and RCH=10 and (MO=3 or MO=6 or MO=9 or MO=12)
```

Nous pouvons simplifier le filtre ci-dessus en utilisant cette fois l'instruction ***IN***

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR>2009 and RCH=10 and MO in (3, 6, 9, 12)
```

Ou pour les mois qui sont autres que 3, 6, 9, 12
c'est à dire : ***NOT IN (3, 6, 9, 12)***

```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR>2009 and RCH=10 and MO NOT IN (3,6,9,12)
```

#### 2.6 Filtre à base d'opérateurs méthématiques

Par exemple, nous pourrions utiliser l'opérateur modulo (%) pour filtrer les mois MO.


```python
%sql SELECT RCH, YR, MO, FLOW_INcms, FLOW_OUTcms From rch WHERE YR>2009 and RCH=10 and MO % 3 = 0
```

### Conclusion du notebook 

Dans l'instruction WHERE, nous pouvons utiliser la combinaison de ***NOT, IN, <>, !=, >=, >, <, <=, AND, OR, ()*** 

et même certains des opérateurs mathématiques (***tels que %, *, /, +, -***) pour récupérer les données souhaitées
