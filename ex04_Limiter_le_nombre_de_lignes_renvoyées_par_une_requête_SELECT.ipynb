# ex04 - Limiter le nombre de lignes renvoyées par une requête SELECT
Souvent, il n'est pas pratique d'afficher toutes les lignes à l'écran, car vous devrez ensuite les faire défiler.... 

Nous pouvons utiliser la clause SQLite "LIMIT" pour restreindre le nombre de lignes renvoyées par une requête. 

Cela peut être utilisé pour tester ou pour éviter de consommer des ressources excessives si la requête renvoie plus d'informations que prévu.

Ce qui suit illustre la syntaxe de la clause LIMIT:
- SELECT column_list FROM table_name LIMIT row_count;
