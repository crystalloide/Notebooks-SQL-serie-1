# ex03 - Récupération de données avec SELECT

Lorsque vous travaillez avec des bases de données et SQL, la tâche la plus courante consiste à demander des données à une ou plusieurs tables, ce qui renvoie ces données sous la forme d'une table de résultats. 

Ces tables de résultats sont appelées result-sets (ensembles de résultats).

L'instruction ***SELECT*** accomplit cela. 

Dans la plupart des applications, ***SELECT*** est la commande de langage de requête de données (DQL) la plus couramment utilisée.

De plus, ***SELECT*** peut faire bien plus que juste récupérer et afficher des données, nous le verrons dans les sections suivantes.

***Syntaxe de la commande SELECT***

- *SELECT colonne1, colonne2, colonneN FROM nom_table;*

Ici : colonne1, colonne2 ... sont les champs (colonnes) de la table (nom_table).

Mais si vous souhaitez interroger tous les enregistrements, utilisez simplement:
- SELECT * FROM nom_table;

### 2. Récupération de toutes les données
Comme nous le savons déjà, l'opération SQL la plus courante est une instruction SELECT, qui extrait des données d'une table, puis affiche les résultats. 

Lançons donc cette requête SQL :

```python
%sql SELECT * From watershed_yearly
```

### 3. Récupération des données pour des colonnes spécifiques

Dans de nombreux cas, il n'est pas nécessaire d'extraire toutes les colonnes lors d'une instruction SELECT. 

Vous pouvez donc sélectionner et choisir uniquement les colonnes qui vous intéressent. 

La requête suivante extrait uniquement les colonnes ***YR***, ***PREC_mm*** et ***PET_mm***:


```python
%sql SELECT YR, PREC_mm, PET_mm From watershed_yearly
```

### 4. Comment faire quelques calculs dans les instructions SELECT

L'instruction SELECT peut faire bien plus que simplement sélectionner des colonnes.

Parfois, nous sommes également intéressés par la relation entre les colonnes. Cela peut être fait avec des expressions dans les instructions SELECT. 

Par exemple, j'aimerais voir la différence entre les colonnes ***PREC_mm*** et ***PET_mm*** (en utilisant l'opérateur moins **-**). 

Vous pouvez également essayer d'autres opérateurs tels que +, *, / ou %.


```python
%sql SELECT YR, PREC_mm, PET_mm, PET_mm-PREC_mm From watershed_yearly
```

Plutôt pratique, mais le résultat est présenté comme un nouveau nom de colonne. 

Comme ça n'est pas très joli, nous allons lui donner un nouveau nom, en utilisant une ***instruction AS (=> un alias)***


```python
%sql SELECT YR, PREC_mm, PET_mm, PET_mm-PREC_mm as PED_mm From watershed_yearly
```

En plus de nommer des expressions à l'aide d'alias, les alias peuvent également être utilisés pour renommer une colonne existante dans la requête. 

Par exemple, nous renommons ***PREC_mm*** en ***Precipitation_mm***.

*** Cependant, une telle opération n'affecte pas les données réelles et ne change pas le nom dans la table***.

Cela ne change finalement ce que vous voyez sur votre écran. 

Vous pouvez toujours essayer la requête "***SELECT * From Watershed_yearly***" pour vous assurer de la chose.


```python
%sql SELECT YR, PREC_mm as Precipitation_mm, PET_mm, PET_mm-PREC_mm as PED_mm From watershed_yearly
```
